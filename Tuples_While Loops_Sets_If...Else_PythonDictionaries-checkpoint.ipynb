```python
# Example 1: Create a Tuple
thistuple = ("apple", "banana", "cherry")
print(thistuple)

# Example 2: Tuple Length
thistuple = ("apple", "banana", "cherry")
print(len(thistuple))

# Example 3: Tuple Items - Data Types
tuple1 = ("apple", "banana", "cherry")
tuple2 = (1, 5, 7, 9, 3)
tuple3 = (True, False, False)
print(tuple1, tuple2, tuple3)

# Example 4: A Tuple with Different Data Types
tuple1 = ("abc", 34, True, 40, "male")
print(tuple1)

# Example 5: type()
mytuple = ("apple", "banana", "cherry")
print(type(mytuple))

# Example 6: The tuple() Constructor
thistuple = tuple(("apple", "banana", "cherry"))
print(thistuple)

# Example 7: Access Tuple Items
thistuple = ("apple", "banana", "cherry")
print(thistuple[1])

# Example 8: Negative Indexing
thistuple = ("apple", "banana", "cherry")
print(thistuple[-1])

# Example 9: Range of Indexes
thistuple = ("apple", "banana", "cherry", "orange", "kiwi", "melon", "mango")
print(thistuple[2:5])

# Example 10: Range of Negative Indexes
thistuple = ("apple", "banana", "cherry", "orange", "kiwi", "melon", "mango")
print(thistuple[-4:-1])

# Example 11: Check if Item Exists
thistuple = ("apple", "banana", "cherry")
if "apple" in thistuple:
    print("Yes, 'apple' is in the fruits tuple")

# Example 12: Change Tuple Values (Convert to List)
x = ("apple", "banana", "cherry")
y = list(x)
y[1] = "kiwi"
x = tuple(y)
print(x)

# Example 13: Add Items (Convert to List)
thistuple = ("apple", "banana", "cherry")
y = list(thistuple)
y.append("orange")
thistuple = tuple(y)
print(thistuple)

# Example 14: Remove Items (Convert to List)
thistuple = ("apple", "banana", "cherry")
y = list(thistuple)
y.remove("apple")
thistuple = tuple(y)
print(thistuple)

# Example 15: Unpacking a Tuple
fruits = ("apple", "banana", "cherry")
(green, yellow, red) = fruits
print(green, yellow, red)

# Example 16: Using Asterisk*
fruits = ("apple", "banana", "cherry", "strawberry", "raspberry")
(green, yellow, *red) = fruits
print(green, yellow, red)

# Example 17: Loop Through a Tuple
thistuple = ("apple", "banana", "cherry")
for x in thistuple:
    print(x)

# Example 18: Loop Through the Index Numbers
thistuple = ("apple", "banana", "cherry")
for i in range(len(thistuple)):
    print(thistuple[i])

# Example 19: Using a While Loop
thistuple = ("apple", "banana", "cherry")
i = 0
while i < len(thistuple):
    print(thistuple[i])
    i += 1

# Example 20: Join Two Tuples
tuple1 = ("a", "b", "c")
tuple2 = (1, 2, 3)
tuple3 = tuple1 + tuple2
print(tuple3)

# Example 21: Multiply Tuples
fruits = ("apple", "banana", "cherry")
mytuple = fruits * 2
print(mytuple)
```

```python
# Example 1: Print i as long as i is less than 6
i = 1
while i < 6:
    print(i)
    i += 1

# Example 2: Break Statement
i = 1
while i < 6:
    print(i)
    if i == 3:
        break
    i += 1

# Example 3: Continue Statement
i = 0
while i < 6:
    i += 1
    if i == 3:
        continue
    print(i)

# Example 4: Else Statement
i = 1
while i < 6:
    print(i)
    i += 1
else:
    print("i is no longer less than 6")
```

```python
# Example 1: If Statement
a = 33
b = 200
if b > a:
    print("b is greater than a")

# Example 2: Elif Statement
a = 33
b = 33
if b > a:
    print("b is greater than a")
elif a == b:
    print("a and b are equal")

# Example 3: Else Statement
a = 200
b = 33
if b > a:
    print("b is greater than a")
elif a == b:
    print("a and b are equal")
else:
    print("a is greater than b")

# Example 4: Short Hand If
a = 200
b = 33
if a > b: print("a is greater than b")

# Example 5: Short Hand If...Else
a = 2
b = 330
print("A") if a > b else print("B")

# Example 6: Multiple Else Statements
a = 330
b = 330
print("A") if a > b else print("=") if a == b else print("B")

# Example 7: And Keyword
a = 200
b = 33
c = 500
if a > b and c > a:
    print("Both conditions are True")

# Example 8: Or Keyword
a = 200
b = 33
c = 500
if a > b or a > c:
    print("At least one of the conditions is True")

# Example 9: Nested If
x = 41
if x > 10:
    print("Above ten,")
    if x > 20:
        print("and also above 20!")
    else:
        print("but not above 20.")

# Example 10: The pass Statement
a = 33
b = 200
if b > a:
    pass
```

```python
# Example 1: Create a Dictionary
thisdict = {
    "brand": "Ford",
    "model": "Mustang",
    "year": 1964
}
print(thisdict)

# Example 2: Access Items
thisdict = {
    "brand": "Ford",
    "model": "Mustang",
    "year": 1964
}
x = thisdict["model"]
print(x)

# Example 3: Get Keys
x = thisdict.keys()
print(x)

# Example 4: Get Values
x = thisdict.values()
print(x)

# Example 5: Get Items
x = thisdict.items()
print(x)

# Example 6: Change Values
thisdict = {
    "brand": "Ford",
    "model": "Mustang",
    "year": 1964
}
thisdict["year"] = 2020
print(thisdict)

# Example 7: Update Dictionary
thisdict = {
    "brand": "Ford",
    "model": "Mustang",
    "year": 1964
}
thisdict.update({"year": 2020})
print(thisdict)

# Example 8: Add Items
thisdict = {
    "brand": "Ford",
    "model": "Mustang",
    "year": 1964
}
thisdict["color"] = "red"
print(thisdict)

# Example 9: Remove Items
thisdict = {
    "brand": "Ford",
    "model": "Mustang",
    "year": 1964
}
thisdict.pop("model")
print(thisdict)

# Example 10: Loop Through a Dictionary
thisdict = {
    "brand": "Ford",
    "model": "Mustang",
    "year": 1964
}
for x in thisdict:
    print(x)

# Example 11: Loop Through Values
thisdict = {
    "brand": "Ford",
    "model": "Mustang",
    "year": 1964
}
for x in thisdict.values():
    print(x)

# Example 12: Loop Through Keys and Values
thisdict = {
    "brand": "Ford",
    "model": "Mustang",
    "year": 1964
}
for x, y in thisdict.items():
    print(x, y)

# Example 13: Copy a Dictionary
thisdict = {
    "brand": "Ford",
    "model": "Mustang",
    "year": 1964
}
mydict = thisdict.copy()
print(mydict)

# Example 14: Nested Dictionaries
myfamily = {
    "child1": {
        "name": "Emil",
        "year": 2004
    },
    "child2": {
        "name": "Tobias",
        "year": 2007
    },
    "child3": {
        "name": "Linus",
        "year": 2011
    }
}
print(myfamily)
```
